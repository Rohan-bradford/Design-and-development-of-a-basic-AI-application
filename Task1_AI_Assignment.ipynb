```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier


#Dataset PATH
dataset_PATH = "CW2_Datasets/dataset1.csv"
DataFrame = pd.read_csv(dataset_PATH)

print("Dataset shape:", DataFrame.shape)
print("\nMissing values:\n", DataFrame.isnull().sum())


X = DataFrame.drop(columns=["id", "level"])
severity_Label = DataFrame["level"].astype(int)
presence_Label = (severity_Label > 0).astype(int)


#Preprocessing
categorical_features = X.select_dtypes(include="object").columns
numerical_features = X.select_dtypes(exclude="object").columns

X[numerical_features] = X[numerical_features].fillna(X[numerical_features].mean())

for i in categorical_features:
    X[i] = X[i].fillna(X[i].mode()[0])

X = pd.get_dummies(X, columns=categorical_features, drop_first=False)

X = X.values.astype(np.float32)
presence_Label = presence_Label.values
severity_Label = severity_Label.values


X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X, presence_Label, test_size=0.2, random_state=42, stratify=presence_Label
)

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X, severity_Label, test_size=0.2, random_state=42, stratify=severity_Label
)


def Model_Evaluation(title, y_true, y_pred, target_names=None):
    print("\n------------------------------")
    print("Model:", title)
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=target_names, zero_division=0))


#Using 3 models for comparison
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
}



print("\n[Binary Classification (0 vs >0)]")
for name, clf in models.items():
    clf.fit(X_train_b, y_train_b)
    preds = clf.predict(X_test_b)
    Model_Evaluation(name, y_test_b, preds, target_names=["No disease", "Disease"])



print("\n[Multiclass Classification (0 to 4)]")
for name, clf in models.items():
    clf.fit(X_train_m, y_train_m)
    preds = clf.predict(X_test_m)
    Model_Evaluation(name, y_test_m, preds)
```
